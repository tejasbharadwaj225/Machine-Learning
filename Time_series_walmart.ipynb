## PERFORMING ANALYSIS FOR STORE 1 DETAILS

# 

```python
df1.head()
```

```python
df1 = df1.set_index(['Date'])
df1
```

```python
df1['Weekly_Sales'].plot()
plt.show()
```

```python
mean_log = df1['Weekly_Sales'].rolling(window = 5).mean()
std_log = df1['Weekly_Sales'].rolling(window = 5).std()
```

```python
mean_log[:20]
```

```python
plt.plot(df1['Weekly_Sales'], color = 'blue')
plt.plot(mean_log, color = 'red' )
plt.plot(std_log, color = 'black')

plt.show()
```

###### Checking whether data is stationary or not by using adfuller test

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df1['Weekly_Sales'])
```

```python
result
```

#####  Hence p value is  less than 0.05, Data is stationary 

###  Observe trend , seasonality and resid by using seasonal_decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose_result = seasonal_decompose(df1['Weekly_Sales'].dropna())
decompose_result.plot()
```

#### ACF (Auto Correlation Function) and PACF(Partial Autocorrelation Function) are used to identify the order of the model in ARIMA models

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

```

```python
acf_plot = acf(df1['Weekly_Sales'].dropna())
plot_acf(acf_plot)
```

```python
pacf_plot = pacf(df1['Weekly_Sales'].dropna())
plot_pacf(pacf_plot, lags = 10)
```

```python
train = df1['Weekly_Sales'].iloc[:120]
test = df1['Weekly_Sales'].iloc[120:]
```

#### Testing With ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train, order = (1,0,1))
model_fit = model.fit()
```

```python
model_fit
```

```python
df1['predict'] = model_fit.predict(start = len(train), end = len(train)+len(test)-1, dynamic = True)
```

```python
df1
```

```python
df1[['Weekly_Sales','predict']].plot()
```

####  Arima model is not able to predict the seasonality 

## SARIMA Model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train, order = (1,0,1), seasonal_order = (1,0,1,5))
model = model.fit()
```

```python
df1['predict'] = model.predict(start = len(train), end = len(train)+len(test)-1, dynamic = True)
```

```python
df1[['Weekly_Sales','predict']].plot()
```

```python
forecasted_results = model.forecast(steps = 36)
```

```python
forecasted_results
```
